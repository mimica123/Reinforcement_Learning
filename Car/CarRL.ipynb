```python
#Import Dependencies
import gym
import highway_env
import numpy as np
from stable_baselines import HER, SAC, PPO2, DQN
```

```python
#Roundabout
env = gym.make('roundabout-v0')
model = PPO2('MlpPolicy', env, verbose=1)
```

```python
#Random Action
for i in range(10):
  done = False
  env.reset()
  while not done:
    #env.render() #comment on colab
    action = env.action_space.sample()
    next_step , reward, done, info = env.step(action)
    print(info)
    print(done)

  env.close()
```

```python
#Create Model
model = PPO2('MlpPolicy', env, verbose=1)
model.learn(total_timesteps=10000)
```

```python
#Save & Load Model
model.save('roundabout')
#del model #activate to save memory
```

```python
model = PPO2.load('roundabout')
```

```python
#Visualize Model
for i in range(10):
  done = False
  obs = env.reset()
  while not done:
    #env.render()
    action, _states = model.predict(obs)
    next_state, reward, done, info = env.step(action)
    print(info)
    print(done)

env.close()
```

```python
#Parking Agent
env = gym.make('parking-v0')
```

```python
#Random Action
for i in range(10):
  done = False
  env.reset()
  while not done:
    #env.render() #comment on colab
    action = env.action_space.sample()
    next_step , reward, done, info = env.step(action)
    print(info)
    print(done)

env.close()
```

```python
model = HER('MlpPolicy', env, SAC, n_sampled_goal=4, goal_selection_strategy='future', verbose=1)
```

```python
model.learn(10000)
```

```python
#Save
model.save('parkingagent')
#del model # active this to erase to save memory
```

```python
#Load
model = HER.load('parkingagent', env=env)
```

```python
#Visualize Model
for i in range(10):
  done = False
  obs = env.reset()
  while not done:
    #env.render()
    action, _states = model.predict(obs)
    next_state, reward, done, info = env.step(action)
    print(info)
    print(done)

env.close()
```
